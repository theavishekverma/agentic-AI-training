```python
import yfinance as yf
import pandas as pd

def check_stock_signal(ticker_symbol):
    # For Indian stocks, add .NS (NSE) or .BO (BSE)
    ticker = yf.Ticker(ticker_symbol)
    
    # Get the last 2 days of data to compare closing vs opening
    data = ticker.history(period="2d")
    
    if len(data) < 2:
        print("Not enough data found for this symbol.")
        return

    # Extract required values
    last_day_close = data['Close'].iloc[-2]  # Closing price of the previous trading day
    today_open = data['Open'].iloc[-1]      # Opening price of today
    current_date = data.index[-1].strftime('%Y-%m-%d')
    
    # Calculations
    diff_amount = today_open - last_day_close
    diff_percent = (diff_amount / last_day_close) * 100
    
    # Trading Logic
    suggestion = "HOLD"
    if today_open < 50:
        suggestion = "BUY (Price below ₹50)"
    elif today_open > 90:
        suggestion = "SELL (Price above ₹90)"
    
    # Display Results
    print(f"{'='*40}")
    print(f"STOCK REPORT FOR: {ticker_symbol}")
    print(f"{'='*40}")
    print(f"Date:                    {current_date}")
    print(f"Last Day Closing Price:  ₹{last_day_close:.2f}")
    print(f"Today Opening Price:     ₹{today_open:.2f}")
    print(f"Difference (Amount):     ₹{diff_amount:.2f}")
    print(f"Difference (Percentage): {diff_percent:.2f}%")
    print(f"Suggestion:              {suggestion}")
    print(f"{'='*40}")

if __name__ == "__main__":
    # Example: TATAMOTORS.NS or RELIANCE.NS
    # Note: Use a low-priced stock like 'SUZLON.NS' to see the 'BUY' trigger
    stock_input = input("Enter Indian Stock Symbol (e.g., SBIN.NS): ").upper()
    check_stock_signal(stock_input)
```
