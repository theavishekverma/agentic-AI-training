```python
import pandas as pd

def read_and_print_csv(file_path):
    try:
        # Load the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Print the entire DataFrame
        print("--- Full CSV Contents ---")
        print(df)

        # Optional: Print just the first 5 rows (useful for large files)
        print("--- First 1 Rows ---")
        print(df.head(1))

        # Optional: Print summary info about columns and data types
        print("--- Summary Information ---")
        print(df.info())

    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    # Replace 'data.csv' with your actual file name
    read_and_print_csv(r"D:\Downloads\Name.csv")
```
