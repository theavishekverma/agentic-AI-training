```python
import pandas as pd
import numpy as np

# 1. Setup Sample Data (Simulating an Indian Stock)
# In a real scenario, you would use: df = pd.read_csv('your_stock_data.csv')
dates = pd.date_range(start='2024-01-01', periods=10, freq='B')
opening_prices = [45.50, 48.20, 52.00, 65.00, 88.50, 92.10, 95.00, 89.00, 47.00, 42.00]

df = pd.DataFrame({
    'Date': dates,
    'Today_Opening_Price': opening_prices
})

# 2. Calculate Last Day Closing Price (Previous day's open used as proxy here)
df['Last_Day_Closing_Price'] = df['Today_Opening_Price'].shift(1)

# 3. Calculate Differences
df['Diff_Amount_INR'] = df['Today_Opening_Price'] - df['Last_Day_Closing_Price']
df['Diff_Percentage'] = (df['Diff_Amount_INR'] / df['Last_Day_Closing_Price']) * 100

# 4. Trading Logic (Buy < 50, Sell > 90)
def get_suggestion(price):
    if price < 50:
        return "BUY"
    elif price > 90:
        return "SELL"
    else:
        return "HOLD"

df['Suggestion'] = df['Today_Opening_Price'].apply(get_suggestion)

# Formatting for clean output
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
df = df.fillna(0) # Handling the first row which has no "Previous Day"

# 5. Print Results Horizontal Column-wise
print(df.to_string(index=False))

# 6. Export to CSV
df.to_csv('D:\\Downloads\\india_stock_results.csv', index=False)
print("\nSuccess: Results exported to 'india_stock_results.csv'")
```
