```python
import textwrap
import requests


def justify_paragraph(text, width=70):
    # Step 1: Clean up any messy whitespace/newlines from the API response
    text = " ".join(text.split())

    # Step 2: Break the text into lines matching our target width
    wrapped_lines = textwrap.wrap(text, width=width)

    justified_lines = []

    # Step 3: Loop through all lines except the very last one
    for line in wrapped_lines[:-1]:
        words = line.split()

        # If it's a single massive word, we can't justify it
        if len(words) == 1:
            justified_lines.append(line)
            continue

        # Calculate how many spaces we need to add to match the target width
        total_chars = sum(len(word) for word in words)
        total_spaces_needed = width - total_chars
        gap_count = len(words) - 1

        # Distribute the spaces evenly between the words
        base_space, extra_space = divmod(total_spaces_needed, gap_count)

        justified_line = ""
        for i, word in enumerate(words[:-1]):
            # Add the base spaces plus any remainder spaces to pad it out
            actual_spaces = base_space + (1 if i < extra_space else 0)
            justified_line += word + (" " * actual_spaces)

        # Append the final word of the line
        justified_line += words[-1]
        justified_lines.append(justified_line)

    # Step 4: The last line of a paragraph shouldn't be stretched out
    if wrapped_lines:
        justified_lines.append(wrapped_lines[-1])

    return "\n".join(justified_lines)


# --- Example API Call ---
url = "https://jsonplaceholder.typicode.com/posts/1"

try:
    response = requests.get(url)
    response.raise_for_status()
    api_data = response.json()

    # Grab a text field from the API response
    raw_api_text = api_data.get("body", "")

    # Generate the justified output (width set to 60 characters)
    print("--- RAW API TEXT ---")
    print(raw_api_text)
    print("\n--- JUSTIFIED PARAGRAPH ---")
    print(justify_paragraph(raw_api_text, width=60))

except Exception as e:
    print(f"An error occurred: {e}")
```
