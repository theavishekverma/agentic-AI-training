```python
import requests

# 1. Define the API endpoint URL
url = "https://api.sampleapis.com/coffee/hot"

try:
    # 2. Send the GET request
    response = requests.get(url)
    print(f"Request sent to {url}. Status Code: {response.status_code}")

    # 3. Check if the request was successful (Status Code 200)
    response.raise_for_status()

    # 4. Parse the data as JSON
    data = response.json()
    
    #print("Success! Data retrieved:")
    #print(f"Title: {data['title']}")
    #print(f"Description: {data['description']}")
    print(response.text)  # Print the raw response text to see the structure of the data
    #print(f"Body: {data['body']}")

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")  # e.g., 404 Not Found or 500 Server Error
except Exception as err:
    print(f"An error occurred: {err}")
```

```python
import requests

# Example URL that returns JSON
url = "https://jsonplaceholder.typicode.com/posts/1"

# Send GET request
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    # Parse JSON response
    data = response.json()
    print("Full JSON response:", data)
    
    # Access specific fields
    print("Title:", data["title"])
    print("Body:", data["body"])
else:
    print("Error:", response.status_code)
```

```python
import requests
import json

url = "https://jsonplaceholder.typicode.com/posts/1"

payload = json.dumps({
  "title": "Title posted by Avishek",
  "body": "This is the body content of my new post by Avishek.",
  "userId": 1,
  "id": 1
})
headers = {
  'Content-Type': 'application/json'
}

response = requests.request("PUT", url, headers=headers, data=payload)

print(response.text)
```
