# Create Charts using MatplotLib

```python
# Bar chart using matplotlib
import matplotlib.pyplot as plt
batter = ['Rahul', 'Virat', 'Rohit', 'Shikhar', 'KL']
runs = [5000, 12000, 9000, 7000, 8000]

#create bar charts
plt.barh(batter, runs, color='red')
plt.xlabel('Batter')
plt.ylabel('Runs')
plt.title('Runs scored by Indian batsmen')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# create Montly sales data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [1000, 1500, 100, 2500, 300, 3500]

# Create Line Chart
plt.plot(months, sales, marker='*', color='blue')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Data')
plt.show()

# Create Bar Charts
plt.bar(months, sales, color='green')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Data')
plt.show()
```

```python
# Stock Price Data

import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib as mpl

# Reset matplotlib configuration
#mpl.rcdefaults()

infy = yf.download('INFY.NS', start='2026-05-01', end='2026-05-16')

# Create line chart instead to show volume data
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(infy.index, infy['Volume'], marker='o', color='orange', linewidth=5, markersize=10)
ax.set_title('INFY Stock Volume')
ax.set_xlabel('Date')
ax.set_ylabel('Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#Mall Customer DataSet

```python
# Create Bar Chart Average Spending Score by Gender
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Mall_Customers.csv')
avg_spending_score = df.groupby('Gender')["Spending Score (1-100)"].mean()
plt.bar(avg_spending_score.index, avg_spending_score.values, color=['blue', 'pink'])
plt.title('Average Spending Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Spending Score')
plt.show()
```

```python
#Customer Spending Trend by Age Group
age_bins = [18, 25, 35, 45, 55, 65]
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
age_group_spending = df.groupby('Age Group')["Spending Score (1-100)"].mean()
plt.bar(age_group_spending.index, age_group_spending.values, color='purple')
plt.title('Customer Spending Trend by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Spending Score')
plt.show()

# Create plot spending Trend by Age Group
plt.plot(age_group_spending.index, age_group_spending.values, marker='o', color='purple')
plt.title('Customer Spending Trend by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Spending Score')
plt.show()
```

```python
#Customer Spending Trend by Annual Income
income_bins = [0, 25, 50, 75, 100, 125, 150]
income_labels = ['0-24k', '25-49k', '50-74k', '75-99k', '100-124k', '125-149k']
df['Income Group'] = pd.cut(df['Annual Income (k$)'], bins=income_bins, labels=income_labels, right=False)
df['Income Group']
income_group_spending = df.groupby('Income Group')["Spending Score (1-100)"].mean()
plt.bar(income_group_spending.index, income_group_spending.values, color='orange')
plt.title('Customer Spending Trend by Annual Income')
plt.xlabel('Annual Income Group')
plt.ylabel('Average Spending Score')
plt.show()


plt.plot(income_group_spending.index, income_group_spending.values,marker='o', color='orange')
plt.title('Customer Spending Trend by Annual Income')
plt.xlabel('Annual Income Group')
plt.ylabel('Average Spending Score')
plt.show()
```
