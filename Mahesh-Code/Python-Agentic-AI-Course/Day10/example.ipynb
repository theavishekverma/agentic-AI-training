```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Tokyo']
}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Perform a simple operation: filter rows where Age > 28
filtered_df = df[df['Age'] > 28]
print(filtered_df)
```

Example of Stock analysis using pandas

```python
import yfinance as yf

ticker = yf.Ticker("TCS.NS")
tcs_data = ticker.history(period="1mo", interval="1d")
print(tcs_data)

if not tcs_data.empty:
    tcs_data['MA20'] = tcs_data['Close'].rolling(window=20, min_periods=1).mean()
    latest = tcs_data.iloc[-1]
    suggestion = "Buy" if latest['Close'] > latest['MA20'] else "Sell"
    print(f"Latest close: {latest['Close']:.2f}")
    print(f"20-day MA: {latest['MA20']:.2f}")
    print(f"Suggestion: {suggestion}")
else:
    print("No data fetched for TCS.")
```
