Mall Customer Data Download

```python
import pandas as pd
df = pd.read_csv('Mall_Customers.csv')
df.isnull().sum()
#df.fillna(0, inplace=True)
df.dropna(inplace=True)
df.isnull().sum()
df.head()
df.groupby('Gender').size()

```

```python

# Create a pivot table showing the average Spending Score by Gender
pivot_table = df.pivot_table(values='Spending Score (1-100)', index='Gender', aggfunc='mean')
print(pivot_table)
```

```python
df.to_json('mall_customers.json', orient='records', lines=True)
```
