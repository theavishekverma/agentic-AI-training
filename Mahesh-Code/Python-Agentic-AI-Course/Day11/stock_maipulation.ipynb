This is stock analysis with different different Technical Indicators

```python
import pandas as pd
import yfinance as yf

infy = yf.download("INFY.NS", start="2025-01-01", end="2026-05-09")
infy.head()
infy.describe()
infy['Profit/Loss'] = infy['Close'] - infy['Open']
infy.tail()
infy['MV_9D'] = infy['Close'].rolling(window=9).mean()
infy['MV_21D'] = infy['Close'].rolling(window=21).mean()

```

```python
# give buy signal when 9D MV crosses above 21D MV and sell signal when 9D MV crosses below 21D MV
infy['Signal'] = ""
infy.loc[infy['MV_9D'] > infy['MV_21D'], 'Signal'] = "Buy"
infy.loc[infy['MV_9D'] < infy['MV_21D'], 'Signal'] = "Sell"
high_price = infy['Close'].max()
low_price = infy['Close'].min()
print(f"Highest closing price: {high_price}")
print(f"Lowest closing price: {low_price}")
high_price=infy['Close']>1800
```
